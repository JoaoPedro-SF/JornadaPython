```python
#importando
import pandas as pd
tabela = pd.read_csv("cancelamentos.csv")
tabela = tabela.dropna()
tabela = tabela.drop("CustomerID", axis=1)
tabela = tabela.drop("assinatura", axis=1)
tabela = tabela.drop("sexo", axis=1)
tabela = tabela.drop("total_gasto", axis=1)
display(tabela)
#Passo 5: Análise profunda da base de dados (encontrando a causa dos cancelamentos)
#importar o plotly para ver as informações em graficos 
import plotly.express as px
# para um grafico funcionar precisa de uma tabela, de um eixo x e um eixo y
#utilizar o tabela.columns para mostrar graficos automaticamente para cada coluna com o For
print(tabela.columns)
for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou")
    grafico.show()
```
